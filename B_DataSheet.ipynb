## Datasheet

### Motivation:
The dataset is hosted by UCI Machine Learning Repository (link below) and the data was donated in the year 2007.
- https://archive.ics.uci.edu/dataset/165/concrete+compressive+strength

The dataset was created by I-Cheng Yeh as referenced in his paper (link below):
- https://www.researchgate.net/publication/222447231_Modeling_of_Strength_of_High-Performance_Concrete_Using_Artificial_Neural_Networks_Cement_and_Concrete_research_2812_1797-1808

Motivation for Data Creation by the Author (quoted from the paper):
- Several studies independently have shown that concrete strength development is determined not only by the water-to-cement ratio, but that it also isinﬂuenced by the content of other concrete ingredients. High-performanceconcrete is a highly complex material, which makes modeling its behavior avery difﬁcult task. This paper is aimed at demonstrating the possibilities ofadapting artiﬁcial neural networks (ANN) to predict the compressive strengthof high-performance concrete.

Motivation for using this data for the project:
- To create a machine learning model which can predict the future strength of a concrete, based on its components and the time for which it is dried.

### Composition:

This file contains 1005 entries for each concrete mixture combinations. This  data is used for this project. 

Data description: The data is numerical.
The meaning of each column and its data type is as below:

- CementComponent (float64): How much cement is mixed
- BlastFurnaceSlag (float64): How much Blast Furnace Slag is mixed
- FlyAshComponent (float64): How much FlyAsh is mixed
- WaterComponent (float64): How much water is mixed
- SuperplasticizerComponent (float64): How much Super plasticizer is mixed
- CoarseAggregateComponent (float64): How much Coarse Aggregate is mixed
- FineAggregateComponent (float64): How much Coarse Aggregate is mixed
- AgeInDays (int64): How many days it was left dry
- Strength (float64): What was the final strength of concrete

### Collection Process
Data Source & Creation (quoted from the paper referenced):
- Experimental data from 17 different sources was used to check the reliability of the strength model. Test data were assembled for concrete containing cement plusﬂy ash, blast furnace slag, and superplasticizer. A determination was made to ensure thatthese mixtures were a fairly representative group governing all of the major parametersthat inﬂuence the strength of HPC and present the complete information required for suchan evaluation


### Preprocessing/cleaning/labelling
No preprocessing, cleaning or labelling was required.
- No Ordinal categorical variable are there that are in string format
- There are no binary nominal variable in string format to be converted

The nominal variable were converetd to numeric using get_dummies() function.
