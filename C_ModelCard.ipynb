## Model Card

### Model Description:
- Input: The data is numerical and the input consists of 8 columns namely; CementComponent, BlastFurnaceSlag, FlyAshComponent, WaterComponent, SuperplasticizerComponent, CoarseAggregateComponent, FineAggregateComponent, AgeInDays.

- Output: Strength of the concrete in MPa as numerical value.

- Model Architecture: MLP regressor model is used. MLP is a powerful tool that uses multilayer perceptron to come up with a best fitted nonlinear line in explaining the relationship between dependent variable and independent variables by learning from the data. The final model was fit on all the input data.

### Performance
- Given the  input as 8 columns namely; CementComponent, BlastFurnaceSlag, FlyAshComponent, WaterComponent, SuperplasticizerComponent, CoarseAggregateComponent, FineAggregateComponent, AgeInDays; The model predicts the concrete strength with an error range of +/- 20Mpa. ### Limitations
- The model has a prediction erro of +/- 20 Mpa. If the concret strength has to be precise and below this error range then this method of prediction will fall short. It is advised that after modelling based on this approach, laboratory testing should be done to verify accuracy of the results.

### Trade-offs
- Data density above Concrete Strength of 55 Mpa is thin and hence the predicted results having strength of >55 Mpa may not be very reliable.
